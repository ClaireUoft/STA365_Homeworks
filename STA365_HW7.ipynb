### Q1: derive the following

1. $p(\boldsymbol \beta | \textbf{y}, \textbf{X}, \Sigma=\sigma^2 I)$ for the **linear regression model** likelihood proportional to $\exp \left(-{\frac {1}{2}}(\mathbf {y} -\mathbf{X\boldsymbol \beta})^\top \Sigma^{-1}(\mathbf {y} - \mathbf{X\boldsymbol\beta})\right)$ and a $\mathcal{MVN}(\boldsymbol\beta_0, \boldsymbol\Sigma_\beta)$ for $\boldsymbol\beta$

Answer:

The linear regression model assumes:

$$
\mathbf{y} = \mathbf{X} \boldsymbol{\beta} + \boldsymbol{\epsilon}, \quad \boldsymbol{\epsilon} \sim \mathcal{N}(0, \sigma^2 I).
$$

The likelihood of ${y}$ given ${\beta}$ is proportional to:

$$
p(\mathbf{y} \mid \mathbf{X}, \boldsymbol{\beta}, \Sigma) \propto \exp \left(-\frac{1}{2}(\mathbf{y}-\mathbf{X} \boldsymbol{\beta})^{\top} \Sigma^{-1}(\mathbf{y}-\mathbf{X} \boldsymbol{\beta})\right)
$$

where $\Sigma=\sigma^2 I$, meaning the errors are i.i.d. normal with variance $\sigma^2$.
The prior on $\boldsymbol{\beta}$ follows a multivariate normal:

$$
p(\boldsymbol{\beta})=\mathcal{M} \mathcal{V} \mathcal{N}\left(\boldsymbol{\beta}_0, \Sigma_\beta\right)
$$

This means that before observing any data, $\boldsymbol{\beta}$ is assumed to be normally distributed around $\boldsymbol{\beta}_0$ with covariance $\Sigma_{\beta}$. The probability density function of this prior is:

$$
p(\boldsymbol{\beta}) \propto \exp \left(-\frac{1}{2}\left(\boldsymbol{\beta}-\boldsymbol{\beta}_0\right)^{\top} \Sigma_\beta^{-1}\left(\boldsymbol{\beta}-\boldsymbol{\beta}_0\right)\right)
$$

By Bayesâ€™ rule, the posterior distribution of ${\beta}$ is proportional to the product of the likelihood and prior:

$$
p(\boldsymbol{\beta} \mid \mathbf{y}, \mathbf{X}, \Sigma) \propto p(\mathbf{y} \mid \mathbf{X}, \boldsymbol{\beta}, \Sigma) p(\boldsymbol{\beta})
$$


Substituting the likelihood and prior:

$$
p(\boldsymbol{\beta} \mid \mathbf{y}, \mathbf{X}, \Sigma) \propto \exp \left(-\frac{1}{2}(\mathbf{y}-\mathbf{X} \boldsymbol{\beta})^{\top} \Sigma^{-1}(\mathbf{y}-\mathbf{X} \boldsymbol{\beta})\right) \times \exp \left(-\frac{1}{2}\left(\boldsymbol{\beta}-\boldsymbol{\beta}_0\right)^{\top} \Sigma_\beta^{-1}\left(\boldsymbol{\beta}-\boldsymbol{\beta}_0\right)\right)
$$


Combining the exponents:

$$
-\frac{1}{2}\left[\boldsymbol{\beta}^{\top}\left(\mathbf{X}^{\top} \Sigma^{-1} \mathbf{X}+\Sigma_\beta^{-1}\right) \boldsymbol{\beta}-2 \boldsymbol{\beta}^{\top}\left(\mathbf{X}^{\top} \Sigma^{-1} \mathbf{y}+\Sigma_\beta^{-1} \boldsymbol{\beta}_0\right)\right]
$$

Recognizing the quadratic form of a multivariate normal density, this matches the quadratic form of a multivariate normal distribution:

$$
\begin{gathered}
\Sigma_{\beta \mid y}=\left(\mathbf{X}^{\top} \Sigma^{-1} \mathbf{X}+\Sigma_\beta^{-1}\right)^{-1} \\
\boldsymbol{\beta} \mid \mathbf{y}, \mathbf{X}, \Sigma \sim \mathcal{M} \mathcal{V} \mathcal{N}\left(\mu_{\beta \mid y}, \Sigma_{\beta \mid y}\right)
\end{gathered}
$$

with parameters:

$$
\Sigma_{\beta \mid y}=\left(\mathbf{X}^{\top} \Sigma^{-1} \mathbf{X}+\Sigma_\beta^{-1}\right)^{-1}
$$
,
$$
\mu_{\beta \mid y}=\Sigma_{\beta \mid y}\left(\mathbf{X}^{\top} \Sigma^{-1} \mathbf{y}+\Sigma_\beta^{-1} \boldsymbol{\beta}_0\right)
$$

This is the posterior distribution of $\beta$ given $\mathbf{y}, \mathbf{X}$, and $\Sigma$.
    

**2.** The analytical posterior distribution for $\sigma^2$ the **error variance** of a **linear regression model** with **design matrix** $\mathbf{X}$ assuming $\sigma^2$ has an **inverse-gamma** prior distribution with parameters $\alpha^*$ and $\beta^*$ (unrelated to $\boldsymbol \beta$).

**Answer:**

We assume the standard linear regression model:

$$
\mathbf{y} = \mathbf{X} \boldsymbol{\beta} + \boldsymbol{\epsilon}, \quad \boldsymbol{\epsilon} \sim \mathcal{MVN}(0, \sigma^2 I)
$$

where $(\mathbf{y}$ is the observed data, $\mathbf{X}$ is the design matrix, $\boldsymbol{\beta}$ is the vector of regression coefficients, and $\sigma^2$ is the unknown error variance.

Since the errors are normally distributed, the likelihood function(probability of data given parameters) follows a multivariate normal distribution:

$$
p\left( \mathbf{y} \mid \mathbf{X}, \boldsymbol{\beta}, \sigma^2 \right) = \mathcal{MVN}(\mathbf{X} \boldsymbol{\beta}, \sigma^2 I)
$$

This has the probability density function:
$$
p\left( \mathbf{y} \mid \mathbf{X}, \boldsymbol{\beta}, \sigma^2 \right) = \propto\left(\sigma^2\right)^{-n / 2} \exp \left(-\frac{1}{2 \sigma^2}(\mathbf{y}-\mathbf{X} \boldsymbol{\beta})^{\top}(\mathbf{y}-\mathbf{X} \boldsymbol{\beta})\right)
$$

We assume that $\sigma^2$ follows an **Inverse-Gamma** prior:

$$
\sigma^2 \sim \operatorname{Inv-Gamma}(\alpha^*, \beta^*)
$$

which has the density function:

$$
p(\sigma^2) \propto (\sigma^2)^{-(\alpha^*+1)} \exp\left(-\frac{\beta^*}{\sigma^2}\right)
$$

By **Bayes' rule**, the **posterior distribution** of $\sigma^2$ is proportional to the product of the likelihood and the prior:

$$
p\left( \sigma^2 \mid \mathbf{y}, \mathbf{X}, \boldsymbol{\beta} \right) \propto p\left( \mathbf{y} \mid \mathbf{X}, \boldsymbol{\beta}, \sigma^2 \right) p\left( \sigma^2 \right)
$$

Substituting the expressions for the likelihood and prior:

$$
p\left( \sigma^2 \mid \mathbf{y}, \mathbf{X}, \boldsymbol{\beta} \right) \propto (\sigma^2)^{-n/2} \exp \left( -\frac{1}{2\sigma^2} (\mathbf{y} - \mathbf{X} \boldsymbol{\beta})^{\top} (\mathbf{y} - \mathbf{X} \boldsymbol{\beta}) \right) 
\times (\sigma^2)^{-(\alpha^*+1)} \exp \left( -\frac{\beta^*}{\sigma^2} \right)
$$

Rearranging:

$$
p\left( \sigma^2 \mid \mathbf{y}, \mathbf{X}, \boldsymbol{\beta} \right) \propto (\sigma^2)^{-(\alpha^* + n/2 + 1)} \exp \left( -\frac{\beta^* + \frac{1}{2} (\mathbf{y} - \mathbf{X} \boldsymbol{\beta})^{\top} (\mathbf{y} - \mathbf{X} \boldsymbol{\beta})}{\sigma^2} \right)
$$

The expression matches the **Inverse-Gamma** density function:

$$
\sigma^2 \sim \operatorname{Inv-Gamma}(\alpha_{\text{new}}, \beta_{\text{new}})
$$

where:

$$
\alpha_{\text{new}} = \alpha^* + \frac{n}{2}
$$

$$
\beta_{\text{new}} = \beta^* + \frac{1}{2} (\mathbf{y} - \mathbf{X} \boldsymbol{\beta})^{\top} (\mathbf{y} - \mathbf{X} \boldsymbol{\beta})
$$

Thus, 

$$
\sigma^2 \mid \mathbf{y}, \mathbf{X}, \boldsymbol{\beta} \sim \operatorname{Inv-Gamma}\left(\alpha^*+\frac{n}{2}, \beta^*+\frac{1}{2}(\mathbf{y}-\mathbf{X} \boldsymbol{\beta})^{\top}(\mathbf{y}-\mathbf{X} \boldsymbol{\beta})\right)
$$

this is the posterior distribution for $\sigma^2$ and is **Inverse-Gamma** with updated parameters.

- For any data set you find interesting (perhaps from kaggle.com?)

```python
import pandas as pd
import numpy as np
import pymc as pm
import arviz as az
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler

# Load dataset
df = pd.read_csv("Sleep_health_and_lifestyle_dataset.csv")

df.columns = df.columns.str.strip()

# Display basic dataset information
print("Dataset Info:\n", df.info())
print("\nSummary Statistics:\n", df.describe())
print("\nFirst Few Rows:\n", df.head())

# Check for missing values
df.fillna(df.mean(numeric_only=True), inplace=True)

# Ensure correct data types
df = df.apply(pd.to_numeric, errors='coerce')

# Introduce synthetic outliers for robust regression
df.loc[0:5, "Sleep Duration"] *= 5  # Artificially inflate some values

# Define features and target variable
X_columns = ["Age", "Physical Activity Level", "Quality of Sleep"]
y_column = "Sleep Duration"

# Extract feature matrix and target vector
X = df[X_columns].values
y = df[y_column].values

# Standardize the features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Add intercept term
X_scaled = np.column_stack([np.ones(X_scaled.shape[0]), X_scaled])  # Add intercept
p = X_scaled.shape[1]

# Bayesian Linear Regression Model
with pm.Model() as model:
    # Priors
    beta0 = pm.Normal("beta0", mu=0, sigma=10)
    betas = pm.Normal("betas", mu=0, sigma=1, shape=p)  # Regression coefficients
    sigma = pm.HalfNormal("sigma", sigma=10)  # Non-inverse-gamma prior for sigma
    
    # Likelihood
    mu = beta0 + pm.math.dot(X_scaled, betas)
    y_obs = pm.Normal("y_obs", mu=mu, sigma=sigma, observed=y)
    
    # Inference
    trace = pm.sample(2000, tune=1000, return_inferencedata=True, target_accept=0.9)
    
# Posterior Analysis
az.plot_trace(trace, var_names=["betas", "sigma", "beta0"])
plt.show()

# Summary of posterior
print(az.summary(trace, var_names=["betas", "sigma", "beta0"]))
```

```python
import pandas as pd
import numpy as np
import pymc as pm
import arviz as az
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler

# Load dataset
df = pd.read_csv("Sleep_health_and_lifestyle_dataset.csv")

df.columns = df.columns.str.strip()

# Display basic dataset information
print("Dataset Info:\n", df.info())
print("\nSummary Statistics:\n", df.describe())
print("\nFirst Few Rows:\n", df.head())

# Check for missing values
df.fillna(df.mean(numeric_only=True), inplace=True)

# Ensure correct data types
df = df.apply(pd.to_numeric, errors='coerce')

# Introduce synthetic outliers for robust regression
df.loc[0:5, "Sleep Duration"] *= 5  # Artificially inflate some values

# Define features and target variable
X_columns = ["Age", "Physical Activity Level", "Quality of Sleep"]
y_column = "Sleep Duration"

# Extract feature matrix and target vector
X = df[X_columns].values
y = df[y_column].values

# Standardize the features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Add intercept term
X_scaled = np.column_stack([np.ones(X_scaled.shape[0]), X_scaled])  # Add intercept
p = X_scaled.shape[1]

# Robust Bayesian Linear Regression Model
with pm.Model() as model:
    # Priors
    betas = pm.Normal("betas", mu=0, sigma=1, shape=p)  # Regression coefficients
    nu = pm.Exponential("nu", 1/10)  # Prior for Student-T degrees of freedom
    sigma = pm.HalfCauchy("sigma", beta=1)  # Non-inverse-gamma prior for sigma
    tau = pm.Gamma("tau", alpha=nu/2, beta=nu/2, shape=len(y))  # Robust precision terms
    
    # Likelihood
    mu = pm.math.dot(X_scaled, betas)
    y_obs = pm.StudentT("y_obs", nu=nu, mu=mu, sigma=sigma * pm.math.sqrt(1/tau), observed=y)
    
    # Inference
    trace = pm.sample(2000, return_inferencedata=True, target_accept=0.9)
    
# Posterior Analysis
az.plot_trace(trace, var_names=["betas", "sigma", "nu"])
plt.show()

# Summary of posterior
print(az.summary(trace, var_names=["betas", "sigma", "nu"]))
```
